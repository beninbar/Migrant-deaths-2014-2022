```python
import pandas as pd
import re
df = pd.read_csv(r'C:\Users\bpinb\OneDrive\Desktop\Coursework\DATA-73200-Interactive-Data-Viz\Final project\Missing_Migrants_Global_Figures.csv')
print(df)
```

```python
coords = df['Coordinates']
print(type(df['Coordinates']))
print(type(df['Coordinates'][0]))
print(len(df['Coordinates']))

nan_in_coords = df[df['Coordinates'].isnull()]
print(len(nan_in_coords))

print(len(df))
dropped = df.dropna(subset=['Coordinates'])
print(len(dropped))
coords = dropped['Coordinates']
print(type(dropped['Coordinates']))
print(len(dropped['Coordinates']))
```

```python
lat = []
long = []
for s in coords:
    x = re.findall('[+-]?\d+\.?\d*', str(s))
    for i, num in enumerate(x):
        if i == 0:
            long.append(num)
        else:
            lat.append(num)

dropped['long'] = long
dropped['lat'] = lat
dropped
```

```python
dropped.to_csv(r'C:\Users\bpinb\OneDrive\Desktop\Coursework\DATA-73200-Interactive-Data-Viz\Final project\Missing_Migrants_processed.csv')
```
